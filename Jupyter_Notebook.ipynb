### In cell 5, the code is supposed to recursively print a string in reverse

```python
def recursiveReverse(str):
    n = len(str) - 1
    if len(str) == 0:
        return ""
    return str[n] + recursiveReverse(str[0:n])

    

print(recursiveReverse("racecar"))
```
